```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
import os 
```

```python
face = []
celeb_name = []
for root, dirs, files in os.walk("C:\Datasets\Face_Img"):
    for folder in dirs:
        for rot, fold, file in os.walk(root+'\\'+folder):
            for img in file:
                celeb_name.append(os.path.basename(folder))
                face.append(root+'\\'+folder+'\\'+img)
celeb_name
```

```python
print(len(face))
```

```python
image_data = []
for data in face:
    image_data.append(cv2.imread(data, cv2.IMREAD_GRAYSCALE))
image_data
```

```python
fig, ax = plt.subplots(3, 5)
for i, axi in enumerate(ax.flat):
    axi.imshow(image_data[i], cmap='bone')
    axi.set(xticks=[], yticks=[], xlabel=celeb_name[i])
```
